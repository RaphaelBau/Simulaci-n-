```python
import numpy as np
import matplotlib.pyplot as plt
```

***Para la simulacion de los Gastos utilizamos una distribucion uniforme***\
 ~$U[30, 50]$

```python
#definimos la funci√≥n para el precio
def uniforme(inf, sup, N):
  l= []
  for i in range (N):
    l.append(np.random.uniform(low=inf, high=sup, size=1))
  return l
```

```python
exp=uniforme(30,50,1000)
```

Se muestran las primeras 10 simulaciones

```python
for j in range(10):
    print('Simulacion ', j+1, '=', exp[j])
```

```python
np.mean(exp)
```

```python
plt.plot(exp)
plt.xlabel('Iteraciones')
plt.ylabel('Gastos simulados')
plt.title('Distribucion Uniforme')
plt.show()
```

```python
def simulacion_uniforme(repeticiones):
    List = []
    for i in range(repeticiones):
        List.append(np.mean(uniforme(30,50,1000)))
    return List
```

```python
simulacion = simulacion_uniforme(1000)

plt.hist(simulacion)
plt.xlabel('iteraciones')
plt.ylabel('p')
plt.title('Probabilidad')
plt.show()
```

***Simulacion de la asistencia del evento***\
Sigue una distribucion normal ~$N[90, 3]$



```python
#Distribucion normal con mu= 90, sigma=3
def normal(mu, sigma, N):
  lista= []
  for i in range (N):
    lista.append(np.random.normal(mu,sigma,1))
  return lista
```

```python
expn= normal(90,3, 1000)
```

```python
for k in range(10):
  print('Simulacion ', k+1, '=', expn[k])
```

```python
np.mean(expn)
```

```python
plt.plot(expn)
plt.xlabel('Iteraciones')
plt.ylabel('Porcentaje de asistencia')
plt.title('Distribucion Normal')
plt.show()
```

```python
def simulacion_normal(repeticiones):
    List = []
    for i in range(repeticiones):
        List.append(np.mean(normal(90,3,1000)))
    return List
```

```python
simulacion = simulacion_normal(1000)

plt.hist(simulacion)
plt.xlabel('iteraciones')
plt.ylabel('p')
plt.title('Probabilidad')
plt.show()
```

Una vez que tenemos la media del porcentaje de los asistentes, calculamos los asistentes que podrian ir al evento


1.   Lugar 2000 personas
2.   Lugar 4800 personas



```python
#total de asistentes al evento
L1=2000
L2=4800

asistentes= np.mean(expn)*(L1+L2)/100
print(asistentes)
```
